```python
import matplotlib.pyplot as plt

%matplotlib inline
```

# functional approach of creating plots

```python
import numpy as np

x = np.linspace(0, 10, 20)
y = x ** 2
```

```python
plt.plot(x, y)
```

## lets add xlable, ylabel and title to plot

```python
plt.plot(x, y)
plt.title('first plot')
plt.xlabel('X')
plt.ylabel('Y')
```

## multi plots

```python
plt.subplot(1, 2, 1) # rows, column, plot number
plt.plot(x, y)

plt.subplot(1, 2, 2) # rows, column, plot number
plt.plot(y, x)
```

# object oriented apporach

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.2, 0.8, 0.8])
ax.plot(x, y)
ax.set_xlabel('X')
```

## subplots

```python
fig, axs = plt.subplots(nrows=2, ncols=2)
axs[0, 1].plot(x, y)
axs[1, 0].plot(y, x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 2), dpi=100)

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
```

# Save figure

```python
fig.savefig('my_fig.png')
```

```python
import matplotlib.image as mpimg

plt.imshow(mpimg.imread('my_fig.png'))
```

# Sample Application for Visulization

## http://paldhous.github.io/ucb/2016/dataviz/data/week3.zip

```python
data = pd.read_csv('nations.csv')
data
```

```python
data['gdp_percap'].fillna(data['gdp_percap'].median(), inplace=True)

data.head(5)
```

```python
data.groupby(['country']).mean()
```

```python
avg_gdp_percap = data.groupby(['country']).mean()['gdp_percap']

avg_gdp_percap
```

```python
top_5_countries = avg_gdp_percap.sort_values(ascending=False).head()
top_5_countries
```

```python
uae = data[data['country']  == 'United Arab Emirates']

uae
```

```python
plt.plot(uae['year'], uae['gdp_percap'])
```

```python
uae.plot.bar(x='year', y='gdp_percap')
```
